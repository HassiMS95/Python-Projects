```python
path = 'words/words.txt'
```

```python
with open(path, 'r') as file:
    wordlist = file.readlines()
```

```python
wordclean = sorted(list(set([word.strip().lower() for word in wordlist])))
```

```python
wordclean[:10]
```

```python
# Task 3
words_bylength = collections.defaultdict(list)
```

```python
for word in wordclean:
    words_bylength[len(word)].append(word)
```

```python
words_bylength
```

```python
words_bysig = collections.defaultdict(list)
```

```python
def signature(word):
    return ''.join(sorted(word))
```

```python
for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
def anagram_fast(word):
    return words_bysig[signature(word)]
```

```python
# Task 4
anagrams_bylength = {length: {word: anagram_fast(word) for word in words_bylength[length] if len(anagram_fast(word)) > 1} for length in words_bylength if len(words_bylength[length]) > 1}
```

```python
anagrams_bylength
```

```python
# Task 5
anagrams_bylength_total = {length: (sum(len(anagram_fast(word)) - 1 for word in words_bylength[length])/2) for length in words_bylength}
```
